```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,1,50) # create a grid from 1 to 50 of 1 point
y = np.log(x) #array containing the values for the natural logarithm function
z = np.exp(x) #array containing the values for the exponential function
plt.scatter(x,y)  # Create scatterplot
plt.scatter(x,z)  # Create scatterplot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,1,50) # create a grid from 1 to 50 of 1 point
y = np.log(x) #array containing the values for the natural logarithm function
z = np.exp(x) #array containing the values for the exponential function
plt.scatter(x,y, label='Natural Logarithm')  # Create scatterplot and lable
plt.scatter(x,z, label='Exponential') # Create scatterplot and lable
plt.xlabel("x") #add label
plt.ylabel("y") #add label
plt.title("Natural Log and Exponential Functions") #title to the graph
plt.legend(loc = 'lower right') #creates a legend in the lower right
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-6.5,6.5,100) # grid of 100 equally spaced points ranging from -6.5 to 6.5
y = np.sin(x) #array containing the values for the natural logarithm function
z = np.cos(x) #array containing the values for the exponential function
plt.plot(x,y, label='Sin')  # Create scatterplot and lable
plt.plot(x,z, label='Cosine') # Create scatterplot and lable
plt.xlabel("x") #add label
plt.ylabel("y") #add label
plt.title("Sine and Cosine Functions") #title to the graph
plt.legend(loc = 'lower left') #creates a legend in the lower left
plt.show()
```
